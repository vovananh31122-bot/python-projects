```python
score = 0 # biến dùng để lưu điểm của người chơi

print("welcome to the quiz game!")
print("You will answer 3 questions.")
print("Good luck!")

# ---------------- QUESTION 1 ----------------
print("\nQuestion 1:")
print("What is the capital of France?")
print("A. London")
print("B. Paris")
print("C. Berlin")

# lấy input từ người chơi và chuyển thành chữ in hoa
# để tránh lỗi người dùng nhập a hoặc A
answer = input("your answer: ").upper()

if answer == "B": # kiểm tra đáp án đúng
  print("correct!!!!")
  score += 1  # nếu đúng thì cộng điểm
else:
  print("wrong:((((")


# ---------------- QUESTION 2 ----------------
print("\nQuestion 2:")
print("What is the capital of Canada?")
print("A. Vancouver")
print("B. Berlin")
print("C. Ottawa")

answer = input("your answer: ").upper()

if answer == "C":
  print("correct!!!!")
  score += 1
else:
  print("wrong:<<<<")


# ---------------- QUESTION 3 ----------------
print("\nQuestion 3:")
print("Which language is used for AI?")
print("A. Python")
print("B. HTML")
print("C. CSS")

answer = input("Your answer: ").upper()

if answer == "A":
    print("correct!!!!")
    score += 1
else:
    print("Wrong!!!")

# in tổng điểm
print("\nYour total score:", score,"/3")

# tính phần trăm điểm
# score = số câu đúng
# 3 = tổng số câu hỏi
percentage = (score / 3) * 100
print("You got", round(percentage, 2), "%")# làm tròn phần trăm đến 2 chữ số

if percentage == 100:# đánh giá kết quả dựa trên phần trăm
    print("Excellent!")
elif percentage >= 50:
    print("Good job!")
else:
    print("Try again!")
```
