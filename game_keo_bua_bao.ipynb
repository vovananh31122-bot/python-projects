```python
import random #cho máy chọn random

choices = ['kéo','búa','bao'] #người chơi và máy chọn random 3 cái này

while True: #vòng lặp vô hạn để game chơi nhiều lần (game loop)
  player = input('mời bạn chọn: kéo, búa, bao: ').lower() #.lower() nghĩa là chuyển tất cả các kiểu player nhập về thường hết
  #chuyển chuỗi thành chữ thường VD: KÉO-->kéo

  #trường hợp player nhập không có trong choices
  if player not in choices:
    print('không hợp lệ! mời chọn lại:')
    continue #quay lại đầu vòng lặp game

  computer = random.choice(choices)

  print('bạn ra:', player)
  print('máy ra:', computer)

  if computer == player:
    print('quề nhau nha!')

  #kiểm tra 3 trường hợp người chơi thắng
  elif  (player == 'búa' and computer == 'kéo') or \
        (player == 'kéo' and computer == 'bao') or \
        (player == 'bao' and computer == 'búa'):
        print('bạn thắng!🎉🎉🎉🎉🎉🎉🎉🎉')

  else:
    print('bạn thua rồi😢😢😢😢😢')

  #hỏi người chơi có muốn chơi lại không
  play_again = input('Bạn có muốn chơi tiếp không? (y/n): ').lower()

  #nếu người chơi không nhập 'y' thì dừng game, 'n' thì tiếp tục
  if play_again != 'y':
      print('thanks for playing!!!!!!!')
      break
```
