```python
import random #random là thư viện tạo số tự nhiên

#lời chào game
print('welcome my litlle game: guess secret number:X?  ^^')
print('tôi là người máy và tôi là người sẽ chọn số bí mật X')
print('gợi ý nhé phạm vi từ 1 đến 100')

#random X
number_x = random.randint(1,100)

#tạo biến số lần đoán
attempts = 0 #số lần người chơi đã đoán
max_attempts = 7 #giới hạn 7 lần đoán


#vòng lặp đoán
while attempts < max_attempts: #chừng nào số lần đoán < 7 → game vẫn tiếp tục
  guess = int(input('mời bạn đoán: '))
  attempts += 1

  if guess < number_x:
    print('😄hơi nhỏ nha, đoán nữa đi: ')
    print('')

  elif guess > number_x:
    print('🙂lố rồi nha, đoán nữaaaa: ')
    print('')

  else: #không lớn hơn cũng không nhỏ hơn nghĩa là bằng
    print('😍đoán zậy mà trúng trời🎉!!!!!!')
    print('số lần đoán：', attempts)
    print("bạn còn", max_attempts - attempts, "lượt")
    break

if guess != number_x: #nếu người chơi lố 7 lượt
  print('hết lượt rồi:)))')
  print('số bí mật là : ', number_x)
  print("Cảm ơn bạn đã chơi game!")



```
